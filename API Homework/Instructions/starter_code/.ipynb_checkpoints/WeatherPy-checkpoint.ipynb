```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time

# Import API key
from config import api_key

# Incorporated citipy to determine city based on latitude and longitude
from citipy import citipy

# Output File (CSV)
output_data_file = "output_data/cities.csv"

# Range of latitudes and longitudes
lat_range = (-90, 90)
lng_range = (-180, 180)
print(api_key)
```

## Generate Cities List

```python
# List for holding lat_lngs and cities
lat_lngs = []
cities = []

# Create a set of random lat and lng combinations
lats = np.random.uniform(low=-90.000, high=90.000, size=1500)
lngs = np.random.uniform(low=-180.000, high=180.000, size=1500)
lat_lngs = zip(lats, lngs)

# Identify nearest city for each lat, lng combination
for lat_lng in lat_lngs:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    
    # If the city is unique, then add it to a our cities list
    if city not in cities:
        cities.append(city)

# Print the city count to confirm sufficient count
len(cities)
cities_pd = pd.DataFrame(cities)
cities_pd.head()
```

```python

#City	Cloudiness	Country	Date	Humidity	Lat	Lng	Max Temp	Wind Speed
cities_pd['Cloudiness'] = ''
cities_pd['Country']= ''
cities_pd['Date'] = ''
cities_pd['Humidity'] = ''
cities_pd['Lat']= ''
cities_pd['Lng']= ''
cities_pd['Max Temp']=''
cities_pd['Wind Speed'] = ''
cities_pd = cities_pd.rename(columns={0: 'City'})
cities_pd.head()
```

### Perform API Calls
* Perform a weather check on each city using a series of successive API calls.
* Include a print log of each city as it'sbeing processed (with the city number and city name).


```python
print("Beginning Data Retrieval")
print("-----------------------------")
params = {"key": api_key}

for index, row in cities_pd.iterrows():
    city = row['City'] 
    base_url = 'http://api.openweathermap.org/data/2.5/weather?q='
    cities_data = requests.get(f'{base_url}{city}&appid={api_key}&units=imperial')
    cities_data = cities_data.json()
    try:
        cities_pd.loc[index,'Cloudiness'] = cities_data['clouds']['all']
        cities_pd.loc[index,'Country'] = cities_data['sys']['country']
        cities_pd.loc[index,'Date'] = cities_data['dt']
        cities_pd.loc[index,'Humidity'] = cities_data['main']['humidity']
        cities_pd.loc[index,'Lat'] = cities_data['coord']['lat']
        cities_pd.loc[index,'Lng'] = cities_data['coord']['lon']
        cities_pd.loc[index,'Max Temp'] = cities_data['main']['temp_max']
        cities_pd.loc[index,'Wind Speed'] = cities_data['wind']['speed']
        print('Processing Record ' + str(index) +' | ' + ' ' + cities_pd.loc[index,'City'])
    except (KeyError, IndexError):
        print("Missing field/result... skipping.")
        
cities_pd.head()   

```

#### Latitude vs. Temperature Plot

```python
x_axis = pd.to_numeric(cities_pd['Lat'])
temp = pd.to_numeric(cities_pd['Max Temp'])
plt.scatter(x_axis,temp,edgecolor = 'black')
plt.title('City Latitude vs. Temperature (9/8/19)')
plt.xlabel('Latitude')
plt.ylabel('Max Temperature (F)')
plt.grid()
plt.plot()
```

#### Latitude vs. Humidity Plot

```python
humidity = pd.to_numeric(cities_pd['Humidity'])
plt.scatter(x_axis,humidity,edgecolor='black')
plt.title('Latitude vs Humidity (9/8/19)')
plt.xlabel('Latitude')
plt.ylabel('Humidity (%)')
plt.ylim(0,110)
plt.grid()
plt.plot()
```

#### Latitude vs. Cloudiness Plot

```python
clouds = pd.to_numeric(cities_pd['Cloudiness'])
plt.scatter(x_axis,clouds,edgecolor='black')
plt.grid()
plt.title('Latitude vs. Cloudiness (9/8/19)')
plt.xlabel('Latitude')
plt.ylabel('Cloudiness (%)')
plt.plot()
```

#### Latitude vs. Wind Speed Plot

```python
wind = pd.to_numeric(cities_pd['Wind Speed'])
plt.scatter(x_axis,wind,edgecolor='black')
plt.grid()
plt.xlabel('Latitude')
plt.ylabel('Wind Speed (mph)')
plt.title('Latitude vs. Winde Speed (9/8/19)')
plt.plot()
```
