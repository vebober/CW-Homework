```python
# Dependencies and Setup
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Hide warning messages in notebook
import warnings
warnings.filterwarnings('ignore')

# File to Load (Remember to Change These)
mouse_drug_data_to_load = "data/mouse_drug_data.csv"
clinical_trial_data_to_load = "data/clinicaltrial_data.csv"

# Read the Mouse and Drug Data and the Clinical Trial Data
mouse_data = pd.read_csv(mouse_drug_data_to_load)
trial_data = pd.read_csv(clinical_trial_data_to_load)

# Combine the data into a single dataset
all_data = pd.merge(mouse_data,trial_data,how = 'inner')

# Display the data table for preview
all_data.head()
```

## Tumor Response to Treatment

```python
# Store the Mean Tumor Volume Data Grouped by Drug and Timepoint 
tumor = all_data.groupby(['Drug','Timepoint'])
average = tumor.mean()['Tumor Volume (mm3)']
# Convert to DataFrame
summary_table1 = pd.DataFrame(average)

# Preview DataFrame
summary_table1.head()

```

```python
# Store the Standard Error of Tumor Volumes Grouped by Drug and Timepoint
from scipy.stats import sem
#summary_table1['Standard Error'] = summary_table1['Tumor Volume (mm3)'].sem()
tumor_standard_errors = tumor.sem()['Tumor Volume (mm3)']

# Convert to DataFrame
summary_table1_standard_errors = pd.DataFrame(tumor_standard_errors)

# Preview DataFrame
summary_table1_standard_errors = summary_table1_standard_errors.unstack(level = ['Drug'])
summary_table1_standard_errors = summary_table1_standard_errors.droplevel(level=0,axis=1)
summary_table1_standard_errors = pd.DataFrame(summary_table1_standard_errors)
summary_table1_standard_errors = summary_table1_standard_errors[['Capomulin','Infubinol',
                                                                'Ketapril','Placebo']]
summary_table1_standard_errors
```

```python
# Minor Data Munging to Re-Format the Data Frames
summary_table1 = summary_table1.unstack(level = ['Drug'])
summary_table1 = summary_table1.droplevel(level=0,axis=1)
summary_table1 = pd.DataFrame(summary_table1)
# Preview that Reformatting worked
summary_table1
```

```python
#Capomulin, Infubinol, Ketapril, and Placebo
summary_table2 = summary_table1[['Capomulin','Infubinol','Ketapril','Placebo']]
summary_table2
```

```python
# Generate the Plot (with Error Bars)
x_axis = summary_table2.index
data = summary_table2.columns
Capomulin = plt.plot(x_axis,summary_table2['Capomulin'],label='Capolumulin',marker='o',color = 'red'
                ,linestyle='dashed',linewidth=1)
Infubinol = plt.plot(x_axis,summary_table2['Infubinol'],label = 'Infubinol',marker='^',color = 'blue',
                linestyle='dashed',linewidth=1)
Ketapril = plt.plot(x_axis,summary_table2['Ketapril'],label = 'Ketapril',marker='s',color = 'green',
                linestyle='dashed',linewidth=1)
Placebo = plt.plot(x_axis,summary_table2['Placebo'],label = 'Placebo',marker='D',color='black',
                linestyle='dashed',linewidth=1)
plt.legend(loc='upper left')
plt.xlabel("Treatment Period (Days)")
plt.ylabel("Tumor Volume(mm3)")
plt.title("Tumor Volume During Treatment")
plt.grid()
plt.errorbar(x_axis,summary_table1_standard_errors)
# Save the Figure
plt.show()
```
